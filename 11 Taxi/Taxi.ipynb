## Обучение

### Подготовка данных

#### Добавление признаков

```python
def make_features(data, max_lag, rolling_mean_size):
    
    data['hour'] = data.index.hour
    data['dayofweek'] = data.index.dayofweek
    

    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    
    train, test = train_test_split(data, shuffle=False, test_size=0.1)
    train = train.dropna()
    
    features_train = train.drop('num_orders',axis=1)
    features_test = test.drop('num_orders',axis=1)
    target_train = train['num_orders']
    target_test = test['num_orders']

    model = LinearRegression()
    rmse = -cross_val_score(model, features_train, target_train, scoring='neg_root_mean_squared_error', cv=TimeSeriesSplit(n_splits=5)).mean()
    print(f'RMSE = {rmse.round(2)} при lag= {max_lag}')
    return data
```

```python
lags_list=[1,3,6,12,24]
data_temp =[]
df_new = df.copy()
for i in range(len(lags_list)):
    data_temp.append(make_features(df_new, lags_list[i], 24))
    display(data_temp[i].head())
```

Минимальное значение RMSE =26,9 при lag= 24, используем это значение параметра для дальнейшего исследования.

```python
rmse_scorer = make_scorer(mean_squared_error, squared=False)

# пайплайн
model_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LinearRegression())
])

tcsv = TimeSeriesSplit(n_splits=5)
param_grid = {
    "lr__normalize": [True, False],
    "lr__copy_X": [True, False],
    "lr__positive": [True, False]
}
gs_lr = GridSearchCV(
    estimator=model_lr,
    param_grid=param_grid,
    cv=tcsv,
    n_jobs=-1,
    scoring=rmse_scorer
)

gs_lr.fit(features_train, target_train)

lr_best_params = gs_lr.best_params_
lr_best_score = gs_lr.best_score_
```

```python
print(lr_best_params)
print(lr_best_score)
```

#### Дерево решений

```python
model_dt= DecisionTreeRegressor(random_state=12345)
param_grid = {
    "max_depth" : range(15,16),
    "min_samples_leaf" : range(20,30)
}
gs_dt = GridSearchCV(model_dt, param_grid = param_grid, scoring=rmse_scorer, cv=TimeSeriesSplit(n_splits=5))
gs_dt.fit(features_train, target_train)
print(f'Лучшая RMSE: {gs_dt.best_score_} при параметрах {gs_dt.best_params_}')
```

####  Градиентный бустинг LightGBM

```python
model_lgbm= LGBMRegressor(random_state=12345, n_jobs=2)

param_grid = {
    "max_depth" : range(11,12),
    "n_estimators" : range(56,57)
}

gs_lgbm = GridSearchCV(model_lgbm, param_grid = param_grid, scoring=rmse_scorer, cv=TimeSeriesSplit(n_splits=5))
gs_lgbm.fit(features_train, target_train)
print(f'Лучшая RMSE: {gs_lgbm.best_score_} при параметрах, {gs_lgbm.best_params_}')
```
